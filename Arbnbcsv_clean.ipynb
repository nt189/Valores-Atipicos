```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargamos el dataframe
df = pd.read_csv('listings_LA_California.csv')
df.head(1)
```

```python
# Creamos un set de las columnas que vamos a utilizar
columnas_permitidas = {'availability_60', 'bedrooms', 'availability_365', 'neighborhood_overview', 'description', 'beds', 'review_scores_accuracy', 'review_scores_checkin', 'host_is_superhost', 'amenities', 'neighbourhood_group_cleansed', 'name', 'review_scores_rating', 'review_scores_communication', 'host_name', 'instant_bookable', 'has_availability', 'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 'host_neighbourhood', 'room_type', 'review_scores_location', 'calculated_host_listings_count', 'accommodates', 'number_of_reviews_l30d', 'price', 'bathrooms_text', 'maximum_nights', 'minimum_nights', 'number_of_reviews_ltm', 'host_response_time', 'host_location', 'host_response_rate', 'host_total_listings_count', 'first_review', 'neighbourhood', 'availability_90', 'neighbourhood_cleansed', 'property_type', 'latitude', 'review_scores_value', 'host_identity_verified', 'host_acceptance_rate', 'number_of_reviews', 'review_scores_cleanliness', 'host_verifications', 'host_listings_count', 'last_review', 'availability_30', 'longitude'}
```

```python
# creamos un dataframe que contenga las columnas que vamos a utlizar
df_col_clean = df.loc[:, df.columns.intersection(columnas_permitidas)]
```

```python
print('Columnas totales',len(df_col_clean.columns))
df_col_clean.head(1)
```

```python
# Hacemos una copia del dataframe con las columnas que vamos a utilizar
df_clean = df_col_clean.copy()
```

```python
df_clean.info()
```

```python
df_clean['price'] = df_clean['price'].str.replace('$', '').str.replace(',', '').astype(float)
df_clean.isnull().sum()
```

```python
#Sustituir valores nulos con promedio o media
df_clean['review_scores_rating'] = df_clean['review_scores_rating'].fillna(round(df['review_scores_rating'].mean(),1))
df_clean['review_scores_accuracy'] = df_clean['review_scores_accuracy'].fillna(round(df['review_scores_accuracy'].mean(),1))
df_clean['review_scores_cleanliness'] = df_clean['review_scores_cleanliness'].fillna(round(df['review_scores_cleanliness'].mean(),1))
df_clean['review_scores_checkin'] = df_clean['review_scores_checkin'].fillna(round(df['review_scores_checkin'].mean(),1))
df_clean['review_scores_communication'] = df_clean['review_scores_communication'].fillna(round(df['review_scores_communication'].mean(),1))
df_clean['review_scores_location'] = df_clean['review_scores_location'].fillna(round(df['review_scores_location'].mean(),1))
df_clean['review_scores_value'] = df_clean['review_scores_value'].fillna(round(df['review_scores_value'].mean(),1))
```

```python
# Sustituir valores nulos con mediana
# Eliminar el sÃ­mbolo de porcentaje y convertir a valores numÃ©ricos
df_clean['host_acceptance_rate'] = df_clean['host_acceptance_rate'].replace('%', '', regex=True).astype(float) / 100
df_clean['host_response_rate'] = df_clean['host_response_rate'].replace('%', '', regex=True).astype(float) / 100
df_clean['price'] = df_clean['price'].replace('%', '', regex=True).astype(float)

# Rellenar los valores nulos con la mediana
df_clean['host_acceptance_rate'] = df_clean['host_acceptance_rate'].fillna(round(df_clean['host_acceptance_rate'].median(), 1))
df_clean['host_response_rate'] = df_clean['host_response_rate'].fillna(round(df_clean['host_response_rate'].median(), 1))
df_clean['price'] = df_clean['price'].fillna(round(df_clean['price'].median(), 1))
```

```python
# Sustituir valores nulos por un string en concreto
df_clean['description'] = df_clean['description'].fillna('Sin descripciÃ³n')
df_clean['neighborhood_overview'] = df_clean['neighborhood_overview'].fillna('Sin descripciÃ³n')
df_clean['first_review'] = df_clean['first_review'].fillna('Sin fecha')
df_clean['last_review'] = df_clean['last_review'].fillna('Sin fecha')
df_clean['host_is_superhost'] = df_clean['host_is_superhost'].fillna('f')
df_clean['host_location'] = df_clean['host_location'].fillna('Mexico')
df_clean['host_neighbourhood'] = df_clean['host_neighbourhood'].fillna('Sin especificar')
df_clean['neighbourhood_group_cleansed'] = df_clean['neighbourhood_group_cleansed'].fillna('Sin descripciÃ³n')
df_clean['neighbourhood'] = df_clean['neighbourhood'].fillna('Sin descripciÃ³n')
df_clean['has_availability'] = df_clean['has_availability'].fillna('f')
```

```python
# sustituir valores nulos por valores no nulos hacia adelante 'forward fill' ('ffill')
df_clean['bathrooms_text'] = df_clean['bathrooms_text'].fillna(method = 'ffill')
df_clean['bedrooms'] = df_clean['bedrooms'].fillna(method = 'ffill')
df_clean['beds'] = df_clean['beds'].fillna(method = 'ffill')
```

```python
# sustituir valores nulos por valores no nulos hacia adelante 'backward fill' ('bfill')
df_clean['bathrooms_text'] = df_clean['bathrooms_text'].fillna(method = 'bfill')
df_clean['host_response_time'] = df_clean['host_response_time'].fillna(method = "bfill")
```
